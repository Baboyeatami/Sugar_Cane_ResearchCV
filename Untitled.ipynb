```python
import numpy as np
import supervision as sv
from ultralytics import YOLO

# ==========================================
# 1. DEFINE PATHS
# ==========================================
yaml_path = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\data.yaml"
# Change these to 'valid' or 'test' depending on what you want to evaluate
eval_images_dir = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\valid\images"
eval_labels_dir = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\valid\labels"
model_weights = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\runs\detect\train\weights\best.pt"

# ==========================================
# 2. SETUP MODEL & DATASET
# ==========================================
# Load YOLOv8 Model
model = YOLO(model_weights)

# Load the Dataset using Supervision
dataset = sv.DetectionDataset.from_yolo(
    images_directory_path=eval_images_dir,
    annotations_directory_path=eval_labels_dir,
    data_yaml_path=yaml_path
)

# ==========================================
# 3. DEFINE INFERENCE CALLBACKS
# ==========================================
# YOLOv8 returns a results list. We take [0] and convert it to Supervision format.
# verbose=False stops YOLO from spamming the console for every single image.
def evaluate_image_cm(image):
    result = model(image, conf=0.5, verbose=False)[0]
    return sv.Detections.from_ultralytics(result)

def evaluate_image_map(image):
    result = model(image, conf=0.01, verbose=False)[0]
    return sv.Detections.from_ultralytics(result)

print("Running evaluation. This may take a moment...\n")

# ==========================================
# 4. COMPUTE METRICS
# ==========================================
map_metric = sv.MeanAveragePrecision.benchmark(
    dataset=dataset,
    callback=evaluate_image_map
)

confusion_matrix = sv.ConfusionMatrix.benchmark(
    dataset=dataset,
    callback=evaluate_image_cm
)

# ==========================================
# 5. PRINT EXACT FORMAT
# ==========================================
print("=======================================================")
print(" 🎯 OVERALL LOCALIZATION PERFORMANCE (mAP)")
print("=======================================================")
print(f"mAP @ 50:      {map_metric.map50 * 100:.2f}%")
print(f"mAP @ 75:      {map_metric.map75 * 100:.2f}%")
print(f"mAP @ 50:95:   {map_metric.map50_95 * 100:.2f}%")
print()

matrix = confusion_matrix.matrix
class_names = dataset.classes

precisions = []
recalls = []
f1_scores = []

print("=======================================================")
print(" 📊 CLASS-WISE CLASSIFICATION (Threshold: 50%)")
print("=======================================================")
print(f"{'Class Name'.ljust(15)} | {'Precision'.ljust(10)} | {'Recall'.ljust(10)} | {'F1-Score'}")
print("-" * 55)

for i, class_name in enumerate(class_names):
    TP = matrix[i, i]
    FP = np.sum(matrix[:, i]) - TP  
    FN = np.sum(matrix[i, :]) - TP  
    
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)
    
    print(f"{class_name.ljust(15)} | {precision*100:>9.2f}% | {recall*100:>9.2f}% | {f1*100:>8.2f}%")

print()
print("=======================================================")
print(" 🏆 OVERALL CLASSIFICATION PERFORMANCE")
print("=======================================================")
avg_precision = np.mean(precisions) * 100
avg_recall = np.mean(recalls) * 100
avg_f1 = np.mean(f1_scores) * 100

print(f"Avg Precision: {avg_precision:.2f}%")
print(f"Avg Recall:    {avg_recall:.2f}%")
print(f"Avg F1-Score:  {avg_f1:.2f}%")
```

```python
import os
# EXPLICITLY HIDE GPU: Must be set before importing torch or model libraries
os.environ["CUDA_VISIBLE_DEVICES"] = "-1" 

import cv2
import glob
import time
import torch
import platform
import psutil
import gc
from tqdm import tqdm

# Model Libraries
from ultralytics import YOLO
from rfdetr import RFDETRNano
import supervision as sv

# ==========================================
# 1. PATHS & SETTINGS
# ==========================================
test_images_dir = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\test\images"

yolo_weights = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\runs\detect\train\weights\best.pt"
rfdetr_weights = r"runs/rfdetr_nano/checkpoint_best_ema.pth"

CONF_THRESH = 0.25 

# Add our explicit class names dictionary here
CLASS_NAMES = {
    0: "Brown-rust",
    1: "Leaf-Scald",
    2: "Red-Rot",
    3: "Smut"
}

output_base = "runs/detect/test_set_renders"
os.makedirs(output_base, exist_ok=True)

yolo_output_dir = os.path.join(output_base, "yolov8_detections")
rf_output_dir = os.path.join(output_base, "rfdetr_nano_detections")

# Get current process for CPU RAM tracking
process = psutil.Process(os.getpid())

# ==========================================
# 2. HARDWARE ENVIRONMENT PROFILING
# ==========================================
print("=======================================================")
print(" 🖥️  HARDWARE & ENVIRONMENT PROFILE (CPU-ONLY MODE)")
print("=======================================================")
print(f"OS:            {platform.system()} {platform.release()}")
print(f"CPU:           {platform.processor()}")
print(f"System RAM:    {psutil.virtual_memory().total / (1024**3):.2f} GB")
print(f"Target Device: CPU (GPU Disabled via Env Var)")
print(f"PyTorch Ver:   {torch.__version__}")
print("=======================================================\n")

print(f"🚀 Mass-Rendering Test Set Images (Threshold: {CONF_THRESH})")

# ==========================================
# 3. RENDER & PROFILE METHOD 1: YOLOv8
# ==========================================
print("\n🎬 Part 1/2: Rendering & Profiling YOLOv8 Detections...")

# Baseline RAM before YOLO
yolo_mem_start = process.memory_info().rss / (1024**2) 

model_yolo = YOLO(yolo_weights)

# YOLO predict natively tracks inference speed
results_yolo = model_yolo.predict(
    source=test_images_dir,
    conf=CONF_THRESH,
    save=True,          
    project=output_base, 
    name="yolov8_detections", 
    exist_ok=True,
    device='cpu' # Forced to CPU
)

yolo_mem_end = process.memory_info().rss / (1024**2)
yolo_peak_cpu_mb = abs(yolo_mem_end - yolo_mem_start) # Model + Inference RAM footprint

yolo_infer_times = [r.speed['inference'] for r in results_yolo]
yolo_avg_ms = sum(yolo_infer_times) / len(yolo_infer_times) if yolo_infer_times else 0
yolo_fps = 1000 / yolo_avg_ms if yolo_avg_ms > 0 else 0
yolo_params = sum(p.numel() for p in model_yolo.model.parameters()) / 1e6 

# Clean up memory before loading the next model to ensure fair RAM testing
del model_yolo
del results_yolo
gc.collect() 

# ==========================================
# 4. RENDER & PROFILE METHOD 2: RF-DETR NANO
# ==========================================
print("\n🎬 Part 2/2: Rendering & Profiling RF-DETR Nano Detections...")
os.makedirs(rf_output_dir, exist_ok=True)

# Baseline RAM before RF-DETR
rf_mem_start = process.memory_info().rss / (1024**2)

model_rf = RFDETRNano(pretrain_weights=rfdetr_weights, num_classes=4)
model_rf.optimize_for_inference() 

box_annotator = sv.BoxAnnotator(thickness=4)
label_annotator = sv.LabelAnnotator(text_thickness=2, text_scale=1.0)

image_paths = glob.glob(os.path.join(test_images_dir, "*.jpg"))
image_paths.extend(glob.glob(os.path.join(test_images_dir, "*.png")))

rf_infer_times = []

if len(image_paths) == 0:
    print(f"❌ Error: No images found in {test_images_dir}")
else:
    for img_path in tqdm(image_paths, desc="Processing RF-DETR Renders"):
        img_name = os.path.basename(img_path)
        
        image = cv2.imread(img_path)
        if image is None:
            continue
            
        # ⏱️ Start Latency Profiling
        start_time = time.perf_counter()
        detections = model_rf.predict(image, threshold=CONF_THRESH)
        end_time = time.perf_counter()
        # ⏱️ End Latency Profiling
        
        rf_infer_times.append((end_time - start_time) * 1000) 
        
        labels = [
            f"{CLASS_NAMES.get(class_id, 'Unknown')} {confidence:0.2f}"
            for class_id, confidence in zip(detections.class_id, detections.confidence)
        ]
        
        annotated_image = image.copy()
        annotated_image = box_annotator.annotate(scene=annotated_image, detections=detections)
        annotated_image = label_annotator.annotate(scene=annotated_image, detections=detections, labels=labels)
        
        save_path = os.path.join(rf_output_dir, img_name)
        cv2.imwrite(save_path, annotated_image)

rf_mem_end = process.memory_info().rss / (1024**2)
rf_peak_cpu_mb = abs(rf_mem_end - rf_mem_start) # Model + Inference RAM footprint

rf_avg_ms = sum(rf_infer_times) / len(rf_infer_times) if rf_infer_times else 0
rf_fps = 1000 / rf_avg_ms if rf_avg_ms > 0 else 0

try:
    if hasattr(model_rf, 'model') and hasattr(model_rf.model, 'parameters'):
        rf_params = sum(p.numel() for p in model_rf.model.parameters()) / 1e6
    elif hasattr(model_rf, 'parameters'):
        rf_params = sum(p.numel() for p in model_rf.parameters()) / 1e6
    else:
        # If wrapped tightly, skip param counting to prevent errors
        rf_params = "Hidden Context"
    
    rf_params_str = f"{rf_params:.2f}" if isinstance(rf_params, float) else str(rf_params)
except:
    rf_params_str = "N/A"

# ==========================================
# 5. IEEE PERFORMANCE REPORT
# ==========================================
print("\n" + "="*75)
print(" 📊 IEEE PAPER: CPU EDGE-SIMULATION BENCHMARK")
print("="*75)
print(f"{'Metric':<30} | {'YOLOv8':<18} | {'RF-DETR Nano':<18}")
print("-" * 75)
print(f"{'Inference Latency (ms/img)':<30} | {yolo_avg_ms:<18.2f} | {rf_avg_ms:<18.2f}")
print(f"{'Frames Per Second (FPS)':<30} | {yolo_fps:<18.2f} | {rf_fps:<18.2f}")
print(f"{'CPU RAM Footprint (MB)':<30} | {yolo_peak_cpu_mb:<18.2f} | {rf_peak_cpu_mb:<18.2f}")
print(f"{'Model Parameters (Millions)':<30} | {yolo_params:<18.2f} | {rf_params_str:<18}")
print(f"{'Testing Batch Size':<30} | {'1':<18} | {'1':<18}")
print("="*75)
print("* Note: Ran explicitly on CPU to simulate edge deployment constraints.")
print(f"✅ Render Complete!")
print("=======================================================")
```

```python
import os
import cv2
import random
import yaml
import supervision as sv
from rfdetr import RFDETRNano

# 1. Load your newly trained EMA weights
model = RFDETRNano(pretrain_weights="runs/rfdetr_nano/checkpoint_best_ema.pth")

# 2. Load your class names directly from your YOLO data.yaml
yaml_path = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\data.yaml"
with open(yaml_path, "r") as f:
    dataset_info = yaml.safe_load(f)
    # The 'names' key in a YOLO yaml holds either a list or a dict of your classes
    class_names = dataset_info["names"] 

# 3. Point to your validation images
val_images_dir = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\valid\images"
all_images = os.listdir(val_images_dir)

# 4. Pick a random image to test
random_image_name = random.choice(all_images)
image_path = os.path.join(val_images_dir, random_image_name)

# 5. Run Inference (Prediction)
image = cv2.imread(image_path)
detections = model.predict(image, threshold=0.4) # Only show boxes with 40%+ confidence

# 6. Build the text labels using the names from the YAML file
labels = [
    f"{class_names[class_id]} {confidence:.2f}"
    for class_id, confidence in zip(detections.class_id, detections.confidence)
]

# 7. Annotate the image
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# Draw the boxes and labels
annotated_image = box_annotator.annotate(scene=image.copy(), detections=detections)
annotated_image = label_annotator.annotate(scene=annotated_image, detections=detections, labels=labels)

# 8. Display the result directly in Jupyter!
sv.plot_image(annotated_image)
```

```python
import os
import numpy as np
import supervision as sv
import matplotlib.pyplot as plt
from rfdetr import RFDETRNano

# 1. Paths to your TEST dataset
yaml_path = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\data.yaml"
test_images_dir = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\valid\images"
test_labels_dir = r"C:\Users\User\Documents\Sugar Cane Disease.v4i.yolov8\valid\labels"
model_weights = "runs/rfdetr_nano/checkpoint_best_ema.pth"

# 2. Load and Optimize the Model
model = RFDETRNano(pretrain_weights=model_weights, num_classes=4)
model.optimize_for_inference()

# 3. Load the TEST Dataset using Supervision
dataset = sv.DetectionDataset.from_yolo(
    images_directory_path=test_images_dir,
    annotations_directory_path=test_labels_dir,
    data_yaml_path=yaml_path
)

# 4. Define Two Inference Callbacks
def evaluate_image_cm(image):
    return model.predict(image, threshold=0.5)

def evaluate_image_map(image):
    return model.predict(image, threshold=0.01)

print("Running evaluation on the Test Set. This may take a moment...\n")

# 5. Compute Localization Metrics (mAP)
print("Calculating Mean Average Precision (Localization)...")
map_metric = sv.MeanAveragePrecision.benchmark(
    dataset=dataset,
    callback=evaluate_image_map
)

# 6. Compute Confusion Matrix (Classification)
print("Calculating Confusion Matrix (Classification)...")
confusion_matrix = sv.ConfusionMatrix.benchmark(
    dataset=dataset,
    callback=evaluate_image_cm
)

# ==========================================
# 7. PRINT LOCALIZATION METRICS (mAP)
# ==========================================
print("\n" + "="*55)
print(" 🎯 OVERALL LOCALIZATION PERFORMANCE (mAP)")
print("="*55)
# These overall metrics are fully supported in your version!
print(f"mAP @ 50:      {map_metric.map50 * 100:.2f}%")
print(f"mAP @ 75:      {map_metric.map75 * 100:.2f}%")
print(f"mAP @ 50:95:   {map_metric.map50_95 * 100:.2f}%")

# ==========================================
# 8. EXTRACT & PRINT CLASSIFICATION METRICS
# ==========================================
matrix = confusion_matrix.matrix
class_names = dataset.classes

precisions = []
recalls = []
f1_scores = []

print("\n" + "="*55)
print(" 📊 CLASS-WISE CLASSIFICATION (Threshold: 50%)")
print("="*55)
print(f"{'Class Name'.ljust(15)} | {'Precision'.ljust(10)} | {'Recall'.ljust(10)} | {'F1-Score'}")
print("-" * 55)

for i, class_name in enumerate(class_names):
    TP = matrix[i, i]
    FP = np.sum(matrix[:, i]) - TP  
    FN = np.sum(matrix[i, :]) - TP  
    
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)
    
    print(f"{class_name.ljust(15)} | {precision*100:>9.2f}% | {recall*100:>9.2f}% | {f1*100:>8.2f}%")

print("\n" + "="*55)
print(" 🏆 OVERALL CLASSIFICATION PERFORMANCE")
print("="*55)
avg_precision = np.mean(precisions) * 100
avg_recall = np.mean(recalls) * 100
avg_f1 = np.mean(f1_scores) * 100

print(f"Avg Precision: {avg_precision:.2f}%")
print(f"Avg Recall:    {avg_recall:.2f}%")
print(f"Avg F1-Score:  {avg_f1:.2f}%")

print("\nDisplaying Confusion Matrix...")

# 9. Plot and display the matrix
confusion_matrix.plot()
```
